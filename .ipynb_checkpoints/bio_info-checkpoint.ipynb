```python
import csv
import numpy as np

data_folder = "C:\\rmin\\phd documents\\bioinformatic\\Data"

file_to_open = data_folder + "\\Dataset1.csv"
class_file_to_open = data_folder + "\\class_dataset1.csv"
print("data file is :")
print(file_to_open)
print("class file is :")
print(class_file_to_open)
```

```python
with open(class_file_to_open,'r') as train_class_file:
    # use list to read csv into array
    class_lines = list(csv.reader(train_class_file,delimiter ='\t'))
    
# first read class result to store into y array
    
array_class_row = class_lines[1]
class_sample_num = len(array_class_row)
y = array_class_row
print(array_class_row)
print(class_sample_num)
print(y[0])
```

```python
# define array string type,  row:feature_number , column 3
# feature_name,min_value,max_value
```

```python

with open(file_to_open,'r') as train_file:
    lines = list(csv.reader(train_file))
    
feature_num = len(lines)-1
array_row = np.array(lines[1])
array_row.shape
sample_num = len(array_row)-1
#sample_num = len(lines[0].split(','))
#print("array_row[0] "+ str(array_row[0]))
print("feature_num " + str(feature_num))
print("sample_num " + str(sample_num))
print("array_row " + str(array_row))
    
```

```python

#data = np.arange(sample_num*feature_num,dtype= np.float64 ).reshape(sample_num,feature_num)
data = np.zeros((sample_num,feature_num))
feature_names=["" for x in range(feature_num)]
for i in range(0,feature_num):
   # get a gene line
    array_string_row = lines[i+1]
    array_np_row = np.array(lines[i+1])
    # get the gene name
    feature_names[i]=str(array_string_row[0])
    for j in range(0,sample_num):
        # get each data of the sample       
        data[j][i] = float(array_np_row[j+1])

print("The first sample data is :"+ str(data[0]))
print(" The first gene name feature is :" + str(feature_names[0]))

print("The second sample data is :"+ str(data[2]))
print(" The second gene name feature is :" + str(feature_names[1]))

# get max_value and min_value of each feature

#[[0 for i in range(cols)] for j in range(rows)] 
config_max_min_array =[["" for y in range (3)]for x in range(feature_num)]

# extract colunm in python, example :[row[0] for row in a] get first column
# numpy.amin() and numpy.amax()

for i in range(0,feature_num):
    array_feature=[row[i] for row in data]
    min_value=np.amin(array_feature)
    max_value=np.amax(array_feature)
    
    # store each feature name, min, max values
    config_max_min_array[i][0]=feature_names[i]
    print("feature name [" + str(i) +"]"+ " is: " + config_max_min_array[i][0])
    config_max_min_array[i][1]=min_value
    print("feature max [" + str(i) +"]"+ " is: " + str(config_max_min_array[i][1]))
    config_max_min_array[i][2]=max_value
    print("feature max [" + str(i) +"]"+ " is: " + str(config_max_min_array[i][2]))
    


train_class_file.close()
train_file.close()
```
